```python
n,l,d=map(int,input().split())
song=[]
#앨범 생성
for i in range(n):
    for j in range(l):
        song.append(1)
    if i!=n-1:
        for k in range(5):
            song.append(0)
cnt=0
#벨소리 생성
ring=[1]
for i in range(d-1):
    ring.append(0)
while len(ring)<=len(song):
    ring=ring+ring
ring=ring+ring

#앨범 소리 0일때 벨소리 1이면 cnt+1하고 출력
for i in range(len(song)):
    if song[i]==0 and ring[i]==1:
        print(i)
        cnt+=1
        break
#앨범끝나고 벨소리 울리는 시간 찾기
if cnt==0:
    while True:
        i+=1
        if ring[i]==1:
            print(i)
            break
        
            
```
