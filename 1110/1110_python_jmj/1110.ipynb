```python
n=int(input())
cnt=0
result=-1
m=n
while n!=result:
    #처음 입력값을 받아오고 10보다 작은 상황을 고려하기위해 썼지만 마음에 들지 않는다
    result=m
    if result<10:
        result=result*10
    a=result//10
    b=result%10
    c=a+b
    #기존값 1의 자리10배 +기존값 각자리 수 더해서 나온수 1의자리
    m=m%10*10+c%10
    #기존값 재설정
    result=m
    #카운팅
    cnt+=1
print(cnt)
        
```
